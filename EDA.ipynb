```python
# first read in the data
dataset=pd.read_csv('train.csv')

# print shape of data with rows and columns
print(dataset.shape)
```

```python
dataset.head()
```

**we will analyze to find out the following

- Missing values
- All the numerical values
- Distribution of the numerical values
- Categorical variables
- Cardinality of categorical variables
- Outliers
- Relationship between independent and dependent features.(sales price)


```python
# finding the features or columns with null values
features_with_na= [features for features in dataset.columns if dataset[features].isnull().sum()>1]

# print out the feature name and the percentage of missing values
for feature in features_with_na:
     print(feature, np.round(dataset[feature].isnull().mean(), 4), '% missing values')
```

## since there are many missing values we need to find out the relationship between the missing values and sales price

lets plot some diagrams to see any relationships

```python
for feature in features_with_na:
    data = dataset.copy()
    # lets make a variable that indicates 1 if the observation was a missing value and zero if it was not.
    data[feature]= np.where(data[feature].isnull(), 1, 0)
    
    # lets calculate the mean sales price where the feature has missing values
    data.groupby(feature)['SalePrice'].median().plot.bar()
    
    plt.title(feature)
    plt.show()
```

From the above we find that missing values have a clear relationship to the dependant value, therefore we will look for a suitable way to replace the values in feature engineering.

## numerical values

```python
numerical_features= [feature for feature in dataset.columns if dataset[feature].dtype !='O']

# printing the number of numerical features
print('number of numerical variables is:', len(numerical_features))
```

```python
# visualize the numerical features
dataset[numerical_features].head()
```

## from the data we can gather that we have date time variables and we will ned to extract data or information from this .

For example by getting the time between when a house was built and the time when it gets sold.

```python
# lets try and get the year variables

year_features = [feature for feature in dataset[numerical_features] if 'Yr'in feature or 'Year' in feature]

year_features
```

```python
for feature in year_features:
    print(feature, dataset[feature].unique())
    
```

```python
# lets try and analyze the relationship between year sold and sales price

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('YrSold')
plt.ylabel('median sale price')
plt.title('House price vs year sold')
```

```python
for feature in year_features:
    if feature != 'YrSold':
        data=dataset.copy()
        data[feature]= data['YrSold']-data[feature]
    
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title('feature vs salesprice')
        plt.show()
```

## discrete numericals

```python
discrete_feature =[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_features and feature not in data['Id']]

discrete_feature
```

```python
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```
